### La librería texttable es para imprimir tablas más bonitas.

```python
import requests
from bs4 import BeautifulSoup
from texttable import Texttable
```

```python
resp = requests.get('https://es.wikipedia.org/wiki/Anexo:Presidentes_de_Colombia')
if resp.status_code != 200:
    print(f'unable to get presidents page. Status Code: {resp.status_code}')
    sys.exit(1)
```

```python
soup = BeautifulSoup(resp.text, 'html.parser')
presidents_tables = soup.find_all('table', {'class': 'wikitable'})
```

## Construcción de Tabla
Primero, se crea una tabla y se añaden algunas configuraciones iniciales sobre el tipo de contenido que va a contener, su cabecera y etc.

```python
pretty_table = Texttable()

pretty_table.set_cols_dtype(['t', 't', 't'])
pretty_table.set_cols_align(['r', 'r', 'r'])
pretty_table.add_row(['Período', 'Nombre del Presidente', 'Partido Político'])
```

Posteriormente, se recorre la tabla de presidentes anteriormente "scrapeada" y por cada elemento, se extraen sus elementos internos: los tr y td, siendo estos últimos los que contienen la información que requerimos.

```python
for presidents_table in presidents_tables:
    presidents = presidents_table.find_all('tr')
    for president in presidents:
        cols = president.find_all('td')
        if len(cols) < 1:
            continue
        
        if cols[1].get_text() == 'Período':
            continue
        
        period = cols[1].get_text()
        name = cols[3].get_text()
        party = cols[4].get_text()
        
        pretty_table.add_row([period, name, party])       
```

## Mostrar Información
Por último, ya con la tabla organizada, se imprime al usuario la información requerida. Esta información contiene: el período de gobernanza, el nombre completo del presidente y su partido político. En algunos casos muy viejos (antes de la república), el partído político está definido por la región que estos gobernaron; después de esta época se pueden encontrar los partidos tradicionales que conocemos.

```python
print(pretty_table.draw())
```
